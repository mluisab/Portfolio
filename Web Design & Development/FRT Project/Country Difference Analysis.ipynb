```python
import pandas as pd
from matplotlib import pyplot as plt

score=pd.read_csv("data-K0Q14.csv")
score.rename(columns = {'X.1':'Country'}, inplace = True)
score= score.drop(columns=["Total Score (inc. COVID)", "COVID"])
df1=score.iloc[:10]
df2=score.iloc[-10:]
final=[df1, df2]
df3=pd.concat(final)
first=df3.plot.bar(x='Country', y='Total Score', rot=90, figsize=(15, 10))
plt.title("Difference between lowest and highest top 10 countries")
```
